```python
import random
from itertools import count
import pandas as pd
import matplotlib
matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

plt.style.use('fivethirtyeight')
plt.ion()
fig, ax = plt.subplots()


x_vals = []
y_vals = []

index = count()


def animate(i):
    data = pd.read_csv('stock_data.csv')
    x= data['Date']
    y1 = data['Price']
    
     
    ax.clear()
    
    ax.plot(x, y1, label='Price')
    
    
    plt.draw()
    plt.pause(0.1)
    
    plt.legend(loc='upper left')
    plt.tight_layout()


ani = FuncAnimation(fig, animate, interval=1000,save_count=1000)

plt.tight_layout()
anim = ani
plt.show()
```
